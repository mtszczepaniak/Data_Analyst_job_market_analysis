```python
def jobs_sumarie(job_postings):
  total_number_of_postings = len(job_postings)
  avg_salary = sum(job['salary'] for job in job_postings) / len(job_postings)
  unique_locations = {job['location'] for job in job_postings}
  return print(f'''
  Liczba ogłoszeń: {total_number_of_postings}
  Średnia pensja: {avg_salary}
  Lokacje: {unique_locations}
  ''')
```
